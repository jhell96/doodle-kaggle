```python
import pandas as pd
import json
import numpy as np
from tqdm import tqdm
from os import listdir
from collections import Counter
from keras.utils import to_categorical
```

```python
indir = "/home/doodle/pedro/data/training_data_grouped/"
```

```python
outdir =  "/home/doodle/pedro/data/training_data_grouped_trimmed/"
```

```python
files = list(listdir(indir))
```

```python
counts = Counter()
```

```python
for file in tqdm(files):
    name = file.split(".")[0]
    counts[name] += len(pd.read_csv(indir+file))
```

```python
json.dump(counts, open("/home/doodle/pedro/data/counts.json", "w"))
```

```python
total = sum(counts.values())
```

```python
samples = {} 
for name, count in counts.items():
    samples[name] = int(count*100000/total)
```

```python
validation = [] 
```

```python
for file in tqdm(files):
    name = file.split(".")[0]
    df = pd.read_csv(indir+file)
    train, validate = np.split(df.sample(frac=1), [len(df)-samples[file.split('.')[0]]])
    validation.append(validate)
    train.to_csv(outdir+file, index=False)
```

```python
validation_df = pd.concat(validation)
```

```python
validation_df.to_csv("/home/doodle/pedro/data/validation.csv", index=False)
```

```python
np.random.seed(seed=0)

class Database:
    def __init__(self):
        indir = "/home/doodle/pedro/data/training_data_grouped/"
        sorted_files = sorted(list(listdir(indir)), key=str.lower) 
        sorted_names = [file.split('.')[0] for file in sorted_files]
        class_counts = json.load(open("/home/doodle/pedro/data/counts.json"))
        total = sum(class_counts.values())
        
        self.streams = [ClassStream(indir+file,y) for y, file in enumerate(sorted_files)] 
        self.prob_dist = [class_counts[name]/total for name in sorted_names] 
        
    def get_next_batch(self, batch_size, prob_dist=None):
        if prob_dist is not None:
            self.prob_dist = prob_dist 
        X, Y = [], [] 
        num_samples = 5
        for i in range(int(batch_size/num_samples)):
            class_stream = np.random.choice(self.streams, p=self.prob_dist)
            for j in range(num_samples):
                x, y = class_stream.get_next() 
                X.append(x)
                Y.append(y)
        return X, Y 
        

class ClassStream:
    def __init__(self, file, y, chunksize=1000):
        self.y = y 
        self.samples = [] 
        self.sample_gen = self.replenish_samples(chunksize) 
        self.file = file 
    
    def get_next(self):
        if len(self.samples) == 0:
            self.samples.extend(next(self.sample_gen))
        return self.samples.pop(), self.y
    
    def replenish_samples(self, chunksize):
        while(True):
            for chunk in pd.read_csv(self.file, chunksize=chunksize):
                yield chunk['drawing']
```

```python
db = Database()
```

```python
for i in tqdm(range(1000)):
    X, Y = db.get_next_batch(1000)
```
